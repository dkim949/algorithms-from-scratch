# 선형 계획법 문제 설명

## 접근 방법
- 문제 모델링: 사과와 오렌지 과일 판매 최적화 문제
- 목적함수: 2x + 3y (최대화)
  - 사과(x) 한 개당 2달러의 이익
  - 오렌지(y) 한 개당 3달러의 이익
  - 총 이익을 최대화하는 것이 목표

## 제약 조건
1. x + y ≤ 10 (총 과일 수량 제약)
   - 가게에서 판매할 수 있는 총 과일 수는 10개로 제한
   - 사과와 오렌지의 합이 10개를 초과할 수 없음

2. 비음수 조건 (x ≥ 0, y ≥ 0)
   - 판매 과일 수량은 음수가 될 수 없음

## 해결 과정
### 가능영역의 꼭짓점 확인
- (0,0): 과일을 전혀 판매하지 않음
- (10,0): 사과만 10개 판매
- (0,10): 오렌지만 10개 판매

### 각 꼭짓점에서 이익 계산
- (0,0): 2×0 + 3×0 = 0달러
- (10,0): 2×10 + 3×0 = 20달러
- (0,10): 2×0 + 3×10 = 30달러

### 최적해 결정
오렌지만 10개 판매할 때 이익 30달러로 최대  
(오렌지의 단위 이익이 더 크므로, 가능한 한 많은 오렌지를 판매)

## 결론
- 최적해: 사과 0개, 오렌지 10개 판매
- 최대 이익: 30달러
- 과일 수량 제약을 완전히 활용 (10/10개, 100%)


```python
import numpy as np
import matplotlib.pyplot as plt

# 문제 정의
# 목적함수: 2x + 3y (이익 최대화)
# 제약조건: x + y ≤ 10 (과일 수량 제약)
# x ≥ 0, y ≥ 0 (비음수 조건)

# 그래프 설정
plt.figure(figsize=(10, 6))

# x, y 범위 설정
x = np.linspace(0, 12, 100)
y = 10 - x  # x + y = 10 직선

# 제약선 그리기
plt.plot(x, y, 'r-', linewidth=2, label='x + y = 10 (과일 수량 제약)')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)  # x축
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)  # y축

# 가능영역 색칠
x_fill = np.linspace(0, 10, 100)
y_fill = 10 - x_fill
plt.fill_between(x_fill, 0, y_fill, color='skyblue', alpha=0.3, label='가능영역')

# 주요 점들 표시 (꼭짓점과 몇 가지 다른 가능한 해)
points = [(0, 0), (10, 0), (0, 10), (5, 5), (2, 8), (8, 2)]
labels = ['(0,0): $0', '(10,0): $20', '(0,10): $30', '(5,5): $25', '(2,8): $28', '(8,2): $22']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown']

for i, (px, py) in enumerate(points):
    plt.scatter(px, py, color=colors[i], s=100, zorder=5)
    plt.text(px+0.3, py+0.3, labels[i], fontsize=10)

# 최적해 강조
plt.scatter(0, 10, color='red', s=150, marker='*', 
            label='최적해: (0,10), 이익: $30', zorder=6)

# 등이익선 그리기
# 이익 30인 등이익선: 2x + 3y = 30
x_iso = np.linspace(0, 15, 100)
y_iso = (30 - 2*x_iso) / 3
plt.plot(x_iso, y_iso, 'b--', linewidth=2, label='등이익선: 2x + 3y = 30')

# 다른 등이익선 추가 (이익 20, 25)
y_iso_20 = (20 - 2*x_iso) / 3
y_iso_25 = (25 - 2*x_iso) / 3
plt.plot(x_iso, y_iso_20, 'g--', linewidth=1, alpha=0.7, label='등이익선: 2x + 3y = 20')
plt.plot(x_iso, y_iso_25, 'y--', linewidth=1, alpha=0.7, label='등이익선: 2x + 3y = 25')

# 그래프 설정
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.xlabel('x (사과 개수)')
plt.ylabel('y (오렌지 개수)')
plt.title('간단한 선형 계획법 문제: 사과와 오렌지 판매')
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# 정수 그리드 추가 (가능한 정수 해 표시)
for i in range(11):
    for j in range(11):
        if i + j <= 10:  # 제약조건 만족
            profit = 2*i + 3*j
            if (i, j) not in points:  # 이미 표시된 주요 점은 제외
                plt.scatter(i, j, color='gray', s=30, alpha=0.5)

plt.show()

# 최적해와 이익 계산
integer_points = [(i, j) for i in range(11) for j in range(11) if i + j <= 10]
profits = [2*i + 3*j for i, j in integer_points]
max_profit_index = np.argmax(profits)
optimal_x, optimal_y = integer_points[max_profit_index]
max_profit = profits[max_profit_index]

print(f"최적해: 사과 {optimal_x}개, 오렌지 {optimal_y}개")
print(f"최대 이익: ${max_profit}")
```

# 접근 방법
## 문제 모델링: 주스 가게의 이익 최대화 문제

## 목적함수: 2x + 3y (최대화)
- 오렌지 주스(x) 한 잔당 2달러의 이익 발생
- 사과 주스(y) 한 잔당 3달러의 이익 발생
- 총 이익을 최대화하는 것이 목표

## 제약 조건
1. x ≤ 5 (오렌지 제약)
   - 오렌지 주스 1잔에 오렌지 2개 필요
   - 오렌지는 10개만 보유 중
   - 따라서 최대 5잔까지만 제조 가능
2. y ≤ 8 (사과 제약)
   - 사과 주스 1잔에 사과 1개 필요
   - 사과는 8개만 보유 중
   - 따라서 최대 8잔까지만 제조 가능
3. x + 0.5y ≤ 5 (시간 제약)
   - 오렌지 주스 1잔 제조에 1시간 소요
   - 사과 주스 1잔 제조에 0.5시간 소요
   - 총 가용 시간은 5시간으로 제한
4. 비음수 조건 (x ≥ 0, y ≥ 0):
   - 주스 생산량은 음수가 될 수 없음

```python
import numpy as np

# 간단한 선형 계획법 문제: 주스 제조 문제
# 목적함수: 2x + 3y (최대화)
# 제약조건: 
# x ≤ 5 (오렌지 제약)
# y ≤ 8 (사과 제약)
# x + 0.5y ≤ 5 (시간 제약)
# x ≥ 0, y ≥ 0 (비음수 조건)

# 제약조건을 y에 대한 함수로 표현
def constraint1(x):
    return np.ones_like(x) * 8  # y ≤ 8

def constraint2(x):
    return (5 - x) * 2  # x + 0.5y ≤ 5 -> y ≤ (5-x)*2

def constraint3(x):
    return np.ones_like(x) * 0  # y ≥ 0 (그냥 x축)

# 가능한 모든 꼭짓점 추출
# (0, 0): 원점
# (5, 0): 오렌지 제약과 x축의 교점
# (0, 8): 사과 제약과 y축의 교점
# (1, 8): 사과 제약과 시간 제약의 교점

vertices = []

# 원점 (0, 0)
vertices.append((0, 0))

# x축과 오렌지 제약의 교점: (5, 0)
vertices.append((5, 0))

# 사과 제약과 y축의 교점: (0, 8)
vertices.append((0, 8))

# 오렌지 제약과 시간 제약의 교점:
# x = 5, x + 0.5y = 5 -> 5 + 0.5y = 5 -> y = 0
# vertices.append((5, 0))  # 이미 추가했지만, 명확하게 다시 표시

# 사과 제약과 시간 제약의 교점:
# y = 8, x + 0.5*8 = 5 -> x + 4 = 5 -> x = 1
vertices.append((1, 8))

print("모든 꼭짓점:", vertices)

# 각 꼭짓점에서 제약조건 만족 여부 확인 및 목적함수 값 계산
feasible_vertices = []
objective_values = []

for x, y in vertices:
    # 제약조건 검사
    orange_constraint = x <= 5
    apple_constraint = y <= 8
    time_constraint = x + 0.5*y <= 5
    x_non_negative = x >= 0
    y_non_negative = y >= 0
    
    if orange_constraint and apple_constraint and time_constraint and x_non_negative and y_non_negative:
        feasible_vertices.append((x, y))
        objective_value = 2*x + 3*y
        objective_values.append(objective_value)
        print(f"꼭짓점 ({x}, {y})는 가능해, 목적함수 값: {objective_value}")
    else:
        constraints = [orange_constraint, apple_constraint, time_constraint, x_non_negative, y_non_negative]
        print(f"꼭짓점 ({x}, {y})는 불가능해, 제약조건 만족 여부: {constraints}")

# 최적해 찾기
if objective_values:
    max_index = np.argmax(objective_values)
    optimal_x, optimal_y = feasible_vertices[max_index]
    optimal_value = objective_values[max_index]
    print("\n최적해:")
    print(f"오렌지 주스(x) = {optimal_x}잔")
    print(f"사과 주스(y) = {optimal_y}잔")
    print(f"최대 이익 = ${optimal_value}")
    
    # 자원 사용량 계산
    oranges_used = 2 * optimal_x
    apples_used = optimal_y
    time_used = optimal_x + 0.5 * optimal_y
    
    print(f"\n자원 사용량:")
    print(f"오렌지: {oranges_used}/10개 ({oranges_used/10*100:.1f}%)")
    print(f"사과: {apples_used}/8개 ({apples_used/8*100:.1f}%)")
    print(f"시간: {time_used}/5시간 ({time_used/5*100:.1f}%)")
else:
    print("가능해가 없습니다.")
```

```python
import numpy as np
from scipy.optimize import linprog

# 문제 데이터
regions = ['A', 'B', 'C']
time_slots = ['점심', '저녁']
orders = {
    ('A', '점심'): 20, ('A', '저녁'): 30,
    ('B', '점심'): 30, ('B', '저녁'): 25,
    ('C', '점심'): 15, ('C', '저녁'): 35
}
base_fee = 3000
incentives = {'A': 500, 'B': 0, 'C': 1000}
total_riders = 20
rider_capacity = 4  # 시간당 처리 건수

# 변수 인덱스 매핑
var_indices = {(r, t): i for i, (r, t) in enumerate([(r, t) for r in regions for t in time_slots])}

# 목적함수: 각 지역/시간대 라이더 한 명이 창출하는 수익
c = []
for r in regions:
    for t in time_slots:
        # 라이더 한 명이 창출하는 수익 (최대화 → 최소화를 위해 음수)
        revenue_per_rider = -(base_fee + incentives[r]) * rider_capacity
        c.append(revenue_per_rider)

# 제약조건 행렬
A_ub = []
b_ub = []

# 1. 총 라이더 수 제약 (각 시간대별)
for t in time_slots:
    constraint = [0] * len(var_indices)
    for r in regions:
        constraint[var_indices[(r, t)]] = 1
    A_ub.append(constraint)
    b_ub.append(total_riders)

# 변수 범위 설정 (x >= 0)
bounds = [(0, None) for _ in range(len(var_indices))]

# 최적화 문제 해결 (주문 처리 제약 제거 - 일부 주문 미처리 허용)
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

print("최적화 성공:", result.success)
print("최적화 메시지:", result.message)

if result.success:
    print("\n최적 라이더 배치:")
    for r in regions:
        for t in time_slots:
            rider_count = result.x[var_indices[(r, t)]]
            print(f"지역 {r}, {t} 시간대: {rider_count:.1f} 명")
    
    # 총 수익 및 미처리 주문 계산
    total_revenue = 0
    total_unprocessed = 0
    
    for r in regions:
        for t in time_slots:
            rider_count = result.x[var_indices[(r, t)]]
            capacity = rider_count * rider_capacity
            processed_orders = min(orders[(r, t)], capacity)
            unprocessed = max(0, orders[(r, t)] - capacity)
            
            revenue = processed_orders * (base_fee + incentives[r])
            total_revenue += revenue
            total_unprocessed += unprocessed
            
            print(f"지역 {r}, {t}: 처리 {processed_orders:.1f}건, 미처리 {unprocessed:.1f}건")
    
    print(f"\n총 수익: {total_revenue:,.0f}원")
    print(f"총 미처리 주문: {total_unprocessed:.1f}건")
```

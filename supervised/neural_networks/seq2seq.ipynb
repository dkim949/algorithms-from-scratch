# Sequence to Sequence (Seq2Seq)

입력 시퀀스를 다른 도메인의 출력 시퀀스로 변환하는 인코더-디코더 구조의 신경망 (예: 영어 문장 → 프랑스어 문장)

## 모델 구조
인코더와 디코더라는 서로 다른 RNN 파트로 구성됨
- 인코더: 입력 시퀀스를 고정 크기 벡터로 압축
- 디코더: 압축된 벡터로부터 출력 시퀀스 생성
- 둘 다 RNN 기반 구조 (주로 LSTM/GRU 사용)

### 순전파 수식

**인코더**:
- $h^{enc}_t = RNN_{enc}(x_t, h^{enc}_{t-1})$
- $c = h^{enc}_T$ (마지막 은닉 상태를 문맥 벡터로 사용)

**디코더**:
- $h^{dec}_t = RNN_{dec}(y_{t-1}, h^{dec}_{t-1}, c)$
- $y_t = softmax(W_y h^{dec}_t + b_y)$

### 구성 요소

- **인코더**:
  - 입력 시퀀스를 순차적으로 처리
  - 전체 입력 정보를 문맥 벡터(context vector)로 압축
  - 양방향 RNN 구조도 가능

- **문맥 벡터**:
  - 인코더의 마지막 은닉 상태
  - 입력 시퀀스의 전체 정보를 담음
  - 인코더와 디코더를 연결하는 브리지 역할

- **디코더**:
  - 문맥 벡터로부터 출력 시퀀스 생성
  - 이전 출력을 다음 입력으로 사용
  - Teacher forcing: 학습 시 실제 정답을 다음 입력으로 사용

## 작동 원리
- encoder의 마지막 h(context vector)를 decoder의 첫 h(초기 은닉 상태)로 사용
- 학습 시엔 teacher forcing, test 시엔 디코더의 출력을 다음 입력으로 사용

### 1. 인코딩 단계
- 입력 시퀀스를 순차적으로 처리
- 시퀀스의 의미를 문맥 벡터로 압축
- 마지막 은닉 상태가 전체 입력 정보 포함

### 2. 디코딩 단계
- <START> 토큰으로 시작
- 이전 출력과 문맥 벡터로 다음 토큰 예측
- <END> 토큰이 나올 때까지 반복

## 주요 특징
- 가변 길이 입출력 처리 가능
- 입출력 시퀀스 길이가 달라도 됨
- 도메인 간 변환 가능 (텍스트↔텍스트, 음성↔텍스트)

## 응용 분야
- 기계 번역
- 텍스트 요약
- 챗봇
- 음성 인식
- 이미지 캡셔닝

## 한계점
- 긴 시퀀스에서 정보 압축의 어려움
- 문맥 벡터의 병목 현상
  - 이를 해결하기 위해 어텐션 메커니즘 도입
- 학습 시간이 오래 걸림
